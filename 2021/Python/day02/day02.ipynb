```python
with open('input.txt') as f:
    commands = tuple(line.strip() for line in f)
```

```python
commands
```

```python
horizontal_position = 0
depth = 0

for command in commands:
    direction, change = command.split()
    change = int(change)
    if direction == 'forward':
        horizontal_position += change
    elif direction == 'down':
        depth += change
    else:
        depth -= change
horizontal_position, depth
```

```python
horizontal_position*depth
```

```python
aim = 0
depth = 0
horizontal_position = 0
```

```python
for command in commands:
    direction, change = command.split()
    change = int(change)
    if direction == 'forward':
        horizontal_position += change
        depth += aim*change
    elif direction == 'down':
        aim -= change
    elif direction == 'up':
        aim += change
    else:
        depth -= change
horizontal_position, depth
```
