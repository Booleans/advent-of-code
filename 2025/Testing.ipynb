```python
from dataclasses import dataclass

@dataclass(frozen=True)
class Rotation:
    direction: str
    num_turns: int

def parse(raw):
    rotations = [] 
    for row in raw:
        direction = row[0]
        num_turns = int(row[1:])
        rotations.append(Rotation(direction, num_turns))
    
    return rotations

with open("inputs/day01.txt") as f:
    raw = f.read().split()

rotations = parse(raw)

dial_position = 50

zero_count = 0
points_at_zero = 0

for rotation in rotations:
    if rotation.direction == 'R':
        dial_position += rotation.num_turns
    else:
        dial_position -= rotation.num_turns
    
    while dial_position not in range(0, 100):
        dial_position = dial_position % 100
        if dial_position != 0:
            points_at_zero += 1

    if dial_position == 0:
        zero_count += 1
        points_at_zero += 1

zero_count, points_at_zero
```
