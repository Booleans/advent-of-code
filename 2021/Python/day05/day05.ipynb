```python
starting_coordinates = []
ending_coordinates = []

with open('input.txt') as f:
    for line in f:
        start, end = line.strip().split(' -> ')
        x1, y1 = start.split(',')
        x2, y2 = end.split(',')
        starting_coordinates.append((int(x1), int(y1)))
        ending_coordinates.append((int(x2), int(y2)))
```
